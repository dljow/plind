```python
edges = np.asfortranarray(np.transpose([[1, 2], [2, 3], [3,4], [4,5], [1,5]]))
points = np.asfortranarray(np.transpose([[2+2j, 1+0j, 3+0j],[1+2j, 2+0j, 3+0j],[1+1j, 7+0j, 4+0j], [1+2j, 1+0j, 4+0j], [3+2j, 3+0j, 1+0j]])) 
bad_edges=[2]
ndim=3
edges_out=np.zeros([2,6], dtype='q')
edges_out.tolist()
points_out= np.append(points, np.zeros([ndim, np.size(bad_edges)]), axis=1)
points_out.tolist()
points_out=np.asfortranarray(points_out)
print(points_out)
edges_out= np.asfortranarray(edges_out)
npts=5
```
